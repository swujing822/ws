```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook"  # 如果无效，试试 "notebook_connected" 或 "iframe_connected"

# df = pd.read_csv('./snapshots/bingx.csv')
df = pd.read_csv('./snapshots/mexc.csv')

print(df.shape)
df.head()
```

```python
df.tail()
```

```python

# 3. 转换时间列为 datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 4. 计算差价
df['spread'] = df['ask'] - df['bid']

# 5. 显示前几条数据确认
# print(df.head())

fig = px.line(
    df.tail(100),
    x='timestamp',
    y='spread',
    color='symbol',
    title='Recent 100 Spread Records',
    labels={'spread': 'Spread (ask - bid)', 'timestamp': 'Time'}
)

fig.update_layout(
    xaxis=dict(rangeslider=dict(visible=True)),
    hovermode='x unified'
)

fig.show()
```
